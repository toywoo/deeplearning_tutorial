### 2.1 퍼셉트론 이란?
다수의 신호의 신호를 입력으로 받아 하나의 신호를 출력한다. 그리고 신호는 1/0 두가지의 값을 가질 수 있다.  
입력 신호가 뉴런에 보내지면 각각의 고유한 가중치가 곱해지고 그 총합이 정해진 한계를 넘어설 때만 1을 출력한다. 그 한계를 임계값이라고 하고 theta로 나타낸다. 

### 2.2 단순한 논리회로
퍼셉트론으로 논리회로를 표현할 수 있는데 게이트 별로 다른 것은 가중치와 임계값 뿐이다. 그리고 기계학습 문제는 이 매개변수 값을 컴퓨터가 자동으로 정하는 것이고 이 과정을 학습이라고 한다.

### 2.3 퍼셉트론 구현하기

```python
# AND 게이트 구현하기
def AND(x1, x2):
    w1, w2, theta = .5, .5, .7
    tmp = x1*w1 + x2*w2
    if tmp <= theta:
        return 0
    elif tmp > theta:
        return 1
```

```python
print(AND(0, 0))
print(AND(1, 0))
print(AND(1, 1))
```

```python
import numpy as np

# 가중치와 편향 구현하기
def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([.5, .5]) # 가중치
    b = -.7                #  편향
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1

def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-.5, -.5]) # 가중치
    b = .7                #  편향
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1

def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([.5, .5]) # 가중치
    b = -.2                #  편향
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1

print(AND(0, 0))
print(AND(1, 0))
print(AND(1, 1))
```

### 2.4 퍼셉트론의 한계
XOR 게이트는 구현할 수가 없다. 퍼셉트론은 직선으로 나뉜 두 영역을 만드는데 XOR 연산을 구별해낼 수 없다. 그래서 직선이 아닌 곡선 비선형 영역으로 표현할 수 있다.

### 2.5 다중 퍼셉트론이 출동한다면
다른 게이트와 조합한다면 가능하지 않을까?
가능하다. 층이 여러 개인 퍼셉트론을 다층 퍼셉트론이라고 한다. 단층 퍼셉트론으로 표현하지 못한 것을 층을 늘려서 구현할 수 있다. 그리고 XOR 같이 비선형적인 표현도 가능하다.

```python
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y

print(XOR(0, 0))
print(XOR(1, 0))
```
