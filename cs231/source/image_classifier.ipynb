```python
import numpy as np
```

```python
class NearestNeighbor:
    
    def __init__(self):
        pass

    # memorize training data
    def train(self, X, y):
        self.Xtr = X
        self.ytr = y

    def predict(self, X):
        num_test = X.shape[0]
        Ypred = np.zeros(num_test, dtype = self.ytr.dtype)

        # for each test img: find closest train img predict label of nearest img, using L1 distance
        for i in xrange(num_test):
            distances = np.sum(np.abs(self.Xtr - X[i:]), axis=1)
            min_index = np.argmin(distances)
            Ypred[i] = self.ytr[min_index]

        return Ypred
```
